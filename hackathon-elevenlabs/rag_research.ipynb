```python
from langchain_community.document_loaders import PyPDFLoader, Docx2txtLoader, UnstructuredExcelLoader, UnstructuredPowerPointLoader

file_path = "C:\\Users\\USER\\Documents\\hackathon\\British National Formulary (BNF85) March to September 2023.pdf"
loader = PyPDFLoader(
    file_path,
    mode="page",
    extraction_mode="layout"
)

docs = loader.load()
```

```python
import pprint

pprint.pp(docs[0].metadata)
```

```python
print(docs[500].page_content)
```

```python

model_kwargs = {'device': 'cuda:0'}

# Cration des embeddings
embedding = HuggingFaceEmbeddings(
    model_name=EMBED_MODEL_ID,
    model_kwargs=model_kwargs
)

# Configuration de Milvus
vector_store_saved = Milvus.from_documents(
    docs,
    embedding,
    collection_name="hackathon",
    connection_args={"uri": "http://localhost:19530"},
)
```

```python
res = vector_store_saved.search(
    query="dapagliflozin side-effects",
    search_type="similarity",
)

for hit in res:
    print(hit.page_content)
```

```python
from typing import Union
from fastapi import FastAPI

app = FastAPI()


@app.get("/")
def read_root():
    return {"Hello": "World"}


@app.get("/query")
def query(q: Union[str, None] = None):
    if q is None:
        return {"result": []}
    res = vector_store_saved.search(
        query=q,
        search_type="similarity",
    )

    return {"result": [hit.page_content for hit in res]}
```
