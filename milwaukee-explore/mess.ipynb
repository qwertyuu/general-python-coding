```python
import pandas as pd

# Load the Excel file
file_path = 'data/Milwaukee Tool Product - March 3.xlsx'
df = pd.read_excel(
    file_path,
    sheet_name='Product Information',
    header=1,
    index_col=None,
)

# Display the first few rows of the dataframe
df.head()
```

```python
# Load the Excel file
df_fr_desc = pd.read_excel(
    'data/Milwaukee Tool Product - March 3.xlsx',
    sheet_name='FR Product Information',
    header=0,
    index_col=None,
)

# Display the first few rows of the dataframe
df_fr_desc.head()
```

```python
only_useful_df = df[["MFG Model # (Series)", "Marketing Copy"]].rename(columns={"MFG Model # (Series)": "model", "Marketing Copy": "description"})
```

```python
only_useful_df[only_useful_df.model == "48-73-6075"]
```

```python
# Load the Excel file
df_price = pd.read_excel(
    'data/01 - Milwaukee IMAP January 2025 PV5.xlsx',
    sheet_name='EVERYDAY IMAP LIST',
    header=1,
    index_col=None,
)

# Display the first few rows of the dataframe
df_price
```

```python
merged_df = df_price.merge(only_useful_df, left_on='ITEM NO.', right_on='model', how="left")
merged_df
```

```python
merged_df = df_fr_desc[["SKU", "Marketing Description French"]].merge(merged_df, left_on='SKU', right_on='model', how="right")
merged_df
```
