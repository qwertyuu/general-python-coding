```python
from transformers import AutoProcessor, AutoModelForVision2Seq
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

model_path = "ibm-granite/granite-vision-3.1-2b-preview"
processor = AutoProcessor.from_pretrained(model_path)
model = AutoModelForVision2Seq.from_pretrained(model_path).to(device)
```

```python

# prepare image and text prompt, using the appropriate prompt template
url = "http://archive.circulaires.com/index.do?flyer=U3VwZXJtYXJjaGUtSUdBLzIwMjUvMDIvMDYvKmZseWVyLTAxLmpwZyoqaW1hZ2UqKmEyNzRkMzVkODk1NzJhMTg1YWQ4MmE1NGE0NjMyNWY5"

conversation = [
    {
        "role": "user",
        "content": [
            {"type": "image", "url": url},
            {"type": "text", "text": "What are all the proteins (meat & fish) on that page? Answer as a json"},
        ],
    },
]
inputs = processor.apply_chat_template(
    conversation,
    add_generation_prompt=True,
    tokenize=True,
    return_dict=True,
    return_tensors="pt"
).to(device)


# autoregressively complete prompt
output = model.generate(**inputs, max_new_tokens=100)
print(processor.decode(output[0], skip_special_tokens=False))
```
